```python
import numpy as np
import pandas as pd
from matplotlib import transforms, pyplot as plt
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

<h1>SVR Feature Importance Graph for Discussion of Results</h1>

```python
#drop unimportant features
kcal_df = pd.read_csv('data/data_clean/kcal_data_clean.csv')
kcal_df = kcal_df.drop(['Country', 'Obesity', 'Undernourished', 'Active', 'Population'], axis=1)

#seperate x and ys
X = kcal_df.drop(['Confirmed','Deaths', 'Recovered'], axis=1)
```

<h3>Confirmed</h3>

```python
svr = SVR(kernel='linear', C=10)
y = kcal_df['Confirmed']
svr.fit(X, y)

# Identifying the most influential features via model coefficients
importance = svr.coef_.ravel()

# Plotting Feature Coefficients
feat_importance = zip(kcal_df.columns, importance)
feat_importance = sorted([x for x in feat_importance], key = lambda x: x[1], reverse=False)

df = pd.DataFrame (feat_importance, columns = ['Feature Name','SVR Coefficient'])
df['positive'] = df['SVR Coefficient'] > 0
ax = df.plot.barh(y="SVR Coefficient", x="Feature Name", color=df.positive.map({True: '#005AB5', False: '#DC3220'}), figsize=(11.25, 7.5))
ax.set_ylabel("Feature Name", fontweight='bold')
plt.title("SVR Coefficients for Predicting Confirmed COVID-19 Cases", fontweight='bold')
ax.set_xlabel("SVR Coefficient", fontweight='bold')
ax.legend([]);
```

<h3>Deaths</h3>

```python
svr = SVR(kernel='linear', C=0.1)
y = kcal_df['Deaths']*10
#X_d = StandardScaler().fit_transform(X)
svr.fit(X, y)

# Identifying the most influential features via model coefficients
importance = svr.coef_.ravel()

# Plotting Feature Coefficients
feat_importance = zip(kcal_df.columns, importance)
feat_importance = sorted([x for x in feat_importance], key = lambda x: x[1], reverse=False)

df = pd.DataFrame (feat_importance, columns = ['Feature Name','SVR Coefficient'])
df['positive'] = df['SVR Coefficient'] > 0
ax = df.plot.barh(y="SVR Coefficient", x="Feature Name", color=df.positive.map({True: '#005AB5', False: '#DC3220'}), figsize=(11.25, 7.5))
ax.set_ylabel("Feature Name", fontweight='bold')
plt.title("SVR Coefficients for Predicting COVID-19 Death Rates", fontweight='bold')
ax.set_xlabel("SVR Coefficient", fontweight='bold')
ax.legend([]);
```
