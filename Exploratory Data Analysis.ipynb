```python
#We want to discover if any of the features are correlated with each other
#import stuff
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#prep protein dataset for eda
protein_df = pd.read_csv("data/Protein_Supply_Quantity_Data.csv")
protein_df= protein_df.replace("<2.5", 2.5)
protein_df = protein_df.dropna()
protein_df = protein_df.drop(['Unit (all except Population)'],axis=1)
#prep fat dataset for eda
fat_df = pd.read_csv("data/Fat_Supply_Quantity_Data.csv")
fat_df = fat_df.replace("<2.5", 2.5)
fat_df = fat_df.dropna()
fat_df = fat_df.drop(['Unit (all except Population)'],axis=1)
#prep food supply kg quantity for eda
kg_df = pd.read_csv("data/Food_Supply_Quantity_kg_Data.csv")
kg_df = kg_df.replace("<2.5", 2.5)
kg_df = kg_df.dropna()
kg_df = kg_df.drop(['Unit (all except Population)'],axis=1)
#prep food supply kcal for eda
kcal_df = pd.read_csv("data/Food_Supply_kcal_Data.csv")
kcal_df = kcal_df.replace("<2.5", 2.5)
kcal_df = kcal_df.dropna()
kcal_df = kcal_df.drop(['Unit (all except Population)'],axis=1)
```

```python
#Now let's investigate the correlation between features and confirmed, deaths, and recovered covid cases.
#Not using active cases, because that is covered by confirmed cases.
```

```python
def diet_correlation(df):
    corr=df.corr()
    corr_final=corr.abs().unstack().sort_values(ascending = False)
    corr_final.drop(corr_final.head(32).index, inplace=True)
    
    print('Diet Features correlation with Confirmed Covid Cases')
    corr_confirmed = corr_final['Confirmed'].head(15)
    corr_confirmed = corr_confirmed.drop(['Recovered', 'Deaths', 'Active', 'Undernourished', 'Obesity'], errors='ignore')
    print(corr_confirmed)
    print()
    
    

    print('Diet Features correlation with Covid Deaths')
    corr_deaths = corr_final['Deaths'].head(15)
    corr_deaths = corr_deaths.drop(['Recovered', 'Confirmed', 'Active', 'Undernourished', 'Obesity'], errors='ignore')
    print(corr_deaths)
    print()
    
    print('Diet Features correlation with Covid Recovered Cases')
    corr_recovered = corr_final['Recovered'].head(14)
    corr_recovered = corr_recovered.drop(['Confirmed', 'Active', 'Deaths', 'Undernourished', 'Obesity'], errors='ignore')
    print(corr_recovered)

    corr_base = corr_deaths + corr_confirmed + corr_recovered
    print()
    print('Combined')
    print(corr_base)
    print('---')
```

```python
print("kg dataset")
diet_correlation(kg_df)
```

```python
#Now we know the strongest 11 correlated features for each confirmed recovered, combined, and deaths 
#for every dataset. Let's make a heatmap using these features for each dataset
```
