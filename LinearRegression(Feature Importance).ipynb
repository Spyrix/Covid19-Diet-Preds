```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ['Deaths', 'Recovered', 'Confirmed']
kcal_cols = [col for col in data.columns if col.endswith('kcal')] + ['Obesity']
data_kcal = data[kcal_cols]
#scaled_feature_names = ['Animal Products_kcal', 'Obesity', 'Vegetal Products_kcal']
# # # log_scaled_feature_names = ['Milk - Excluding Butter_kcal']


# data_kcal[scaled_feature_names] = StandardScaler().fit_transform(data_kcal[scaled_feature_names])
 
# X = data_kcal
# X = StandardScaler().fit_transform(X)
# y = data['Confirmed']
# X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, shuffle = True, random_state = 42)
```

## Using kcal dataset

#### Linear Regression for kcal dataset on Confirmed Cases

```python
# Separating X and y and scaling X due to data distribution
X = data_kcal
X = MinMaxScaler().fit_transform(X)
y = data['Confirmed']

# Train-test split of data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, shuffle = True, random_state = 42)

# Instantiate and fit model
lr_confirmed = LinearRegression()
lr_confirmed.fit(X_train, y_train)

print('Mean squared error: %.2f'
      % mean_squared_error(y_test, lr_confirmed.predict(X_test)))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, lr_confirmed.predict(X_test)))
```
