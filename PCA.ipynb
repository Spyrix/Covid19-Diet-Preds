```python
#import stuff
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
#prep protein dataset for pca
protein_df = pd.read_csv("data/Protein_Supply_Quantity_Data.csv")
protein_df= protein_df.replace("<2.5", 2.5)
protein_df = protein_df.dropna()
protein_df = protein_df.drop(['Unit (all except Population)'],axis=1)
#prep fat dataset for pca
fat_df = pd.read_csv("data/Fat_Supply_Quantity_Data.csv")
fat_df = fat_df.replace("<2.5", 2.5)
fat_df = fat_df.dropna()
fat_df = fat_df.drop(['Unit (all except Population)'],axis=1)
#prep food supply kg quantity for pca
kg_df = pd.read_csv("data/Food_Supply_Quantity_kg_Data.csv")
kg_df = kg_df.replace("<2.5", 2.5)
kg_df = kg_df.dropna()
kg_df = kg_df.drop(['Unit (all except Population)'],axis=1)
#prep food supply kcal for pca
kcal_df = pd.read_csv("data/Food_Supply_kcal_Data.csv")
kcal_df = kcal_df.replace("<2.5", 2.5)
kcal_df = kcal_df.dropna()
kcal_df = kcal_df.drop(['Unit (all except Population)'],axis=1)
#prep merged_df for pca
merged_df = pd.read_csv("data/merged_df.csv")
```

```python
def PCA_function(dataset):
    #designate the 2nd to 26th columns as our X
    X = dataset.iloc[:, 1:25].values
    #designate 
    y = dataset.iloc[:, 26].values

    #extract the column names
    column_names=dataset.columns[1:25]

    X = StandardScaler().fit_transform(X)

    # Splitting the dataset into the Training set and Test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

    X_train_ori = X_train
    X_test_ori = X_test
    
    most_important = []
    # Applying PCA
    for i in range (5,0,-1):
        pca = PCA(n_components = i)
        X_train = pca.fit_transform(X_train_ori)
        X_test = pca.transform(X_test_ori)
        explained_variance = pca.explained_variance_ratio_
        print(explained_variance)
        n_pcs= pca.components_.shape[0]
        loading_scores = [np.abs(pca.components_[i]).argmax() for i in range(n_pcs)]
        most_important_names = [column_names[loading_scores[i]] for i in range(n_pcs)]
        #print(most_important)
        print(most_important_names)
        print("------")
    
    X_new = X[:,loading_scores[0]]
    y_new = y

    axes = plt.gca()
    #axes.set_xlim([min(X_new[:])-.15,max(X_new[:])+.15])
    #axes.set_ylim([min(y_new[:])-.09,max(y_new[:])+.001])
    axes.spines["bottom"].set_color("purple")
    axes.spines["left"].set_color("purple")
    axes.tick_params(axis='x', colors='purple')
    axes.tick_params(axis='y', colors='purple')

    for i in range (0,X.shape[0]):
        plt.scatter(X_new[i], y_new[i], s = 100, c = 'red')
        if i%2 == 1:
            plt.annotate(dataset.iloc[i, 0], (X_new[i], y_new[i]), fontsize=16,rotation=-45,va='top')
        else:
            plt.annotate(dataset.iloc[i, 0], (X_new[i], y_new[i]), fontsize=16,rotation=+45,va='bottom')

    plt.title(column_names[loading_scores[0]] +' Consumed vs Confirmed cases',fontsize=20, fontweight='bold',c = 'purple')
    plt.xlabel(column_names[loading_scores[0]] + ' Consumed Percentage',fontsize=16, fontweight='bold',c = 'purple')
    plt.ylabel('Confirmed Cases Percentage',fontsize=16, fontweight='bold',c = 'purple')

    figure = plt.gcf()  # get current figure
    figure.set_size_inches(32, 18) # set figure's size manually to your full screen (32x18)
    plt.show()
```

```python
PCA_function(protein_df)
```

```python
PCA_function(fat_df)
```

```python
PCA_function(kg_df)
```

```python
PCA_function(kcal_df)
```
